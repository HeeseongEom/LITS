```python
import json
import glob
import os
ct_data_dir = "D:/severance/imagesTr"
rt_data_dir = "D:/severance/labelsTr"

# 원하는 데이터 수
num_data_training = 353

# training 데이터를 담을 리스트 초기화
training_data = []

# CT 데이터와 RT 데이터를 디렉토리에서 동적으로 불러와서 리스트에 추가
ct_file_paths = glob.glob(os.path.join(ct_data_dir, "*.nii.gz"))
rt_file_paths = glob.glob(os.path.join(rt_data_dir, "*.nii.gz"))

# 파일 이름에서 숫자만 추출하여 정렬
ct_file_paths.sort(key=lambda x: int(''.join(filter(str.isdigit, os.path.basename(x)))))
rt_file_paths.sort(key=lambda x: int(''.join(filter(str.isdigit, os.path.basename(x)))))

# 데이터 수를 원하는 수로 제한
ct_file_paths = ct_file_paths[:num_data_training]
rt_file_paths = rt_file_paths[:num_data_training]

for ct_path, rt_path in zip(ct_file_paths, rt_file_paths):
    training_data.append({"image": os.path.join("imagesTr", os.path.basename(ct_path)), "label": os.path.join("labelsTr", os.path.basename(rt_path))})

# 기존 "training" 키의 값을 업데이트
my_data["training"] = training_data
```

```python
num_data_validation = 80

# training 데이터를 담을 리스트 초기화
training_data = []

# CT 데이터와 RT 데이터를 디렉토리에서 동적으로 불러와서 리스트에 추가
ct_file_paths = glob.glob(os.path.join(ct_data_dir, "*.nii.gz"))
rt_file_paths = glob.glob(os.path.join(rt_data_dir, "*.nii.gz"))

# 파일 이름에서 숫자만 추출하여 정렬
ct_file_paths.sort(key=lambda x: int(''.join(filter(str.isdigit, os.path.basename(x)))))
rt_file_paths.sort(key=lambda x: int(''.join(filter(str.isdigit, os.path.basename(x)))))

# 데이터 수를 원하는 수로 제한
ct_file_paths_training = ct_file_paths[:num_data_training]
rt_file_paths_training = rt_file_paths[:num_data_training]

for ct_path, rt_path in zip(ct_file_paths_training, rt_file_paths_training):
    training_data.append({"image": os.path.join("imagesTr", os.path.basename(ct_path)), "label": os.path.join("labelsTr", os.path.basename(rt_path))})

# 기존 "training" 키의 값을 업데이트
my_data["training"] = training_data

# validation 데이터를 담을 리스트 초기화
validation_data = []

# CT 데이터와 RT 데이터를 디렉토리에서 동적으로 불러와서 리스트에 추가
ct_file_paths_validation = ct_file_paths[num_data_training:num_data_training + num_data_validation]
rt_file_paths_validation = rt_file_paths[num_data_training:num_data_training + num_data_validation]

for ct_path, rt_path in zip(ct_file_paths_validation, rt_file_paths_validation):
    validation_data.append({"image": os.path.join("imagesTr", os.path.basename(ct_path)), "label": os.path.join("labelsTr", os.path.basename(rt_path))})

# 기존 "validation" 키의 값을 업데이트
my_data["validation"] = validation_data
```
